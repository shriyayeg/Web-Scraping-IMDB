```python
import csv

soup = BeautifulSoup(imbd, 'html.parser')

# li[contains(@class,'ipc-metadata-list-summary-item')]

movies = soup.find_all("li", class_="ipc-metadata-list-summary-item")

movie_data_list = []
for movie in movies:
    title = movie.find('h3', class_='ipc-title__text').text.strip()
    ranking = re.sub(r'\..*','',title)
    title = re.sub(r'.*\. ','',title)
    elements = movie.find_all('span', class_='dli-title-metadata-item')
    rating_and_votes = movie.find('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating').text.strip() if movie.find('span', class_='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating') else 'n/a'
    rating = re.sub(r'\(.*','', rating_and_votes)
    votes_element = movie.find('span', class_='ipc-rating-star--voteCount')
    votes = votes_element.text.strip() if votes_element else 'n/a'
    votes = re.sub(r'\((.*?)\)', r'\1', votes)
    metascore_element = movie.find('span', class_='metacritic-score-box')
    metascore = metascore_element.text.strip() if metascore_element else 'n/a'

    # getting the year, length, and rating from element in the html
    elements_data = [element.get_text(strip=True) for element in elements]
    
    # append movie data to the list
    movie_data_list.append([ranking, title] + elements_data + [rating, votes, metascore])

# csv_file = 'movies_data.csv'
file_path = "/Users/hill3/Downloads/movie_data.csv"

# Write the data into a CSV file
with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # writing headers for the data
    writer.writerow(['Ranking', 'Title', 'Year', 'Runtime', 'Rating', 'Review Ratings', 'Votes', 'Metascore'])
    writer.writerows(movie_data_list)

print("CSV file created successfully.")
```
